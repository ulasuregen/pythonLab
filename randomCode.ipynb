```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
row_ann = pd.DataFrame({'kmeans':['C' + str(i) for i in np.random.randint(3, size = 5)], 
'hc': ['D' + str(i) for i in np.random.randint(3, size = 5)],
'bla': ['E' + str(i) for i in np.random.randint(5, size = 5)]})
```

```python
unique_row = np.unique(row_ann)
print(row_ann)
print(unique_row)
```

```python
palette = sns.color_palette('husl', len(unique_row))
print(palette)
```

```python
color_mapping = dict(zip(unique_row, palette))
color_mapping
```

```python
row_ann[['kmeans','hc']].map(lambda x : color_mapping[x])
```
