```python
def climbingLeaderboard(ranked, player):
    ranked = sorted(list(set(ranked)), reverse= True)
    print(ranked)
    print(player)
    result = []

    for p in player:
        start = 0
        end = len(ranked) - 1
        
        if(p >= ranked[start]): 
            result.append(1)
            continue
        if(p < ranked[end]):   
            result.append(end + 2)
            continue
    
        mid = (start + end) // 2

        while True:
            if(p == ranked[mid]):
                result.append(mid + 1)
                break
            elif(p > ranked[mid]):
                end = mid
            else:
                start = mid 
            
            mid = (start + end) // 2

            if(start+1 == end):
                result.append(start + 2)
                break
            
    return result


    
```

```python
def banzhaf_indices(weights : dict, quota):
    from itertools import combinations # we will use this function which just generates combinations
    weights = {k : v for k,v in weights.items() if isinstance(v,(int,float))} # Removes unwanted data there I just look if it is int or float
    players = list(weights.keys()) # Get the names of the voters
    critical_elections = {p : 0 for p in players} # Each time the vote of p_i changes the election result, we increment its score by 1 and store it in this dictionary


    for i in range(len(players)): # The subsets sizes can be between 0 and (#players - 1); here i represent the size of a subset 
        for subset in combinations(weights, i): # we generate i sized subsets from candidates.
            sum_subset = sum(weights[v] for v in subset) # if the sum has already reached the quota, then we don't need to calculate anything further, I added it for performance
            
            if sum_subset < quota:
                looked_player =  set(players) - set(subset) # we just will look the players, which are not in the subset. 
                
                for lp in looked_player: 
                    if sum_subset + weights[lp] >= quota: # if the candidates vote makes a difference, then we increment its critical_election number by 1 
                        critical_elections[lp] += 1

    total_power = sum(v for v in critical_elections.values())

    if total_power != 0:
        critical_elections = {k : round(v / total_power,2) for k, v in critical_elections.items()} # here we normalize the power indices, so that they are all non-negative and sum to 1
    else:
        critical_elections = {k : 1 / len(weights.items()) for k, v in critical_elections.items()}

    return critical_elections            
    
```
