# Question To Myself

What if I have two movies and a bunch of participants rated those movies. Then I would have N x 2 matrix. 

Probably some people won’t have watched both of the movies, so some ratings will be missing. 
Now the column rank could be at most two, since we have only two columns, and therefore row rank is also at most 2.

If I select two independent rows, then I should be able to fill the gaps with linear approximation. Right ? 

```python
import numpy as np
import pandas as pd
from plotnine import *
from itertools import combinations
```

```python
N = 50

np.random.seed(42)
df = pd.DataFrame(
    np.random.randint(1, 11, size = (N,2)), 
    columns= ['StarWars', 'Inception']
)

df.head()
```

Now with probability p, I will delete the entry in our matrix

```python
p = 0.3

mask = np.random.rand(*df.shape) < p
df_nan = df.mask(mask)

df_nan.isna().sum()
```

Now we will find two linearly independent row vectors:

```python
clean = df_nan.dropna() # we will drop the rows which contains a NaN

base1 = None
base2 = None

for i in combinations(clean.index, 2):
    v1 = clean.loc[i[0]]
    v2 = clean.loc[i[1]]

    determinant = v1[0]*v2[1] - v1[1] * v2[0]
    
    if determinant != 0:
        if base1 is None:
            base1 = [i[0], i[1]]
        elif base2 is None:
            base2 = [i[0], i[1]]
            break
```

Now we have 2 different bases for the range of rows

```python
rows_with_one_nan = df_nan[df_nan.isna().sum(axis=1) == 1]
rows_with_one_nan.head()
```
